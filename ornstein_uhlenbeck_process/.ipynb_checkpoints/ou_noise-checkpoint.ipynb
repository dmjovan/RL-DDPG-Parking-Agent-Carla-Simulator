## Importing necessary libriaries

```python
import numpy as np
import matplotlib.pyplot as plt
import os 

np.random.seed(1)
```

## Ornstein-Uhlenbeck process

```python
def ornstein_uhlenbeck_noise(mu, sigma, theta, dt, tmax, x0):
    
    t = np.linspace(0,tmax,int(1.0/dt))
    
    x = np.zeros_like(t)
    
    x[0] = x0
    
    for i in range(t.shape[0]-1):
        x[i+1]= x[i] + theta*(mu-x[i])*dt + sigma*np.sqrt(dt)*np.random.normal(size=1)
    
    return t, x
```

## Simulation of OU process

```python
fig = plt.figure(figsize = (12,10), dpi = 120)

for sigma, line_color in zip([0.1, 0.2, 0.3, 0.4], ['blue', 'green', 'tomato', 'purple']):
    t, x = ornstein_uhlenbeck_noise(0.0,sigma,0.15,0.01,1000000,0.0)
    plt.plot(t,x, color = line_color, label='$\\sigma = $' + str(sigma))

plt.axhline(0.0, color='red', linewidth=2.0, label='mean')
plt.title('Ornstein-Uhlenbeck noise with different $\\sigma$ ($\\mu='+ str(0.0) +', \\theta='+ str(0.15) +')$')
plt.xlabel('t [s]')
plt.ylabel('noise(t)')
plt.legend(loc='upper left')
plt.grid()

plt.show()
plt.tight_layout()
fig.savefig(os.getcwd()+'/ou_action_noise.png', facecolor = 'white', bbox_inches='tight')
```
