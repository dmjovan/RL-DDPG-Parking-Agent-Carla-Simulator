```python
import numpy as np
import matplotlib.pyplot as plt
from  matplotlib import patches
from matplotlib.figure import Figure
from matplotlib import rcParams
from scipy.fft import fft, ifft
import scipy.signal as signal
import scipy.io as sio
import scipy.io.wavfile as wav
from IPython.display import Audio
import math
from array import array
```

```python
ecg = sio.loadmat('ecg_corrupted.mat')

ecg_signal = ecg['ecg_corrupted'].squeeze()

fs = 360
Ts = 1 / fs
t = np.arange(start = 0, stop = len(ecg_signal) * Ts, step = Ts)

fig, ax = plt.subplots()
ax.plot(t, ecg_signal)
ax.set(xlabel='t [s]', ylabel='EKG [mV]',
       title='Originalni EKG signal')
```

```python
# BASELINE DRIFT FILTER

# fs - ucestanost odabiranja
# fa - granicna ucestanost nepropusnog opsega
# fp - granicna ucestanost propusnog opsega
# Aa, Ap - odgovarajuca slabljenja u nepropusnom/propusnom opsegu
# b - brojilac
# a - imenilac

def baseline_drift_filter(fs, fa, fp, Aa, Ap):
    T = 1 / fs
    # transformacija f[Hz] -> w[rad/s] i predistorzija ucestanosti
    wa = 2 / T * math.tan(2*math.pi*fa/fs)
    wp = 2 / T * math.tan(2*math.pi*fp/fs)
    # za normalizovani NF filtar
    waN = 1
    wpN = wa / wp
    
    # faktor selektivnosti
    k = wpN / waN
    # faktor diskriminacije
    D = (math.pow(10, (0.1 * Aa)) - 1) / (math.pow(10, (0.1 * Ap)) - 1)
    # red funkcije prenosa
    N = math.ceil(math.acosh(math.sqrt(D)) / math.acosh(1/k))
    
    flag = 0
    
    while flag == 0:
        # odredjivanje nula i polova funkcije prenosa normalizovanog NF filtra
        z, p, g = signal.cheb2ap(N, Aa)
        
        # odredjivanje koeficijenata polinoma na osnovu nula i polova
        bn = np.poly(z) * g
        an = np.poly(p)
        
        fh = 2*math.pi*fa
        # NF -> VF
        bh, ah = signal.lp2hp(bn, an, fh)
        # bilinearna transformacija
        # preslikavanje iz analognog u digitalni domen
        b, a = signal.bilinear(bh, ah, fs)
        # racunanje amplitudske karakteristike
        f, Hd = signal.freqz(b, a, worN=15000, fs=fs)
        hd = abs(Hd)
        
        # provera gabarita
        if (20 * math.log10(max(hd[f<fa])) < -Aa) and (20 * math.log10(min(hd[f>fp])) > -Ap):
            flag = 1
        else:
            N = N + 1
        
    return b, a
```

```python
# projektovanje filtra pomocu baseline_drift_filter funkcije
fab = 0.4
fpb = 1
Aab = 30
Apb = 0.5
bb, ab = baseline_drift_filter(fs, fab, fpb, Aab, Apb)
ecg_baseline_filtered = signal.lfilter(bb, ab, ecg_signal)
# vremenski oblik dobijenog signala
fig, ax = plt.subplots()
ax.plot(t, ecg_baseline_filtered)
ax.set(xlabel='t [s]', ylabel='EKG [mV]',
       title='Parcijalno filtrirani EKG signal')
```

```python
# POWERLINE NOISE FILTER

# fs - ucestanost odabiranja
# fc - centralna ucestanost (60Hz za primenu u projektu)
# Aa/Ap - odgovarajuca slabljenja u nepropusnom/propusnom opsegu
# b = brojilac
# a = imenilac

def power_line_noise_filter(fs, fc, Aa, Ap):
    # propusni opseg
    fp1 = (fc - 2)
    fp2 = (fc + 2)
    # nepropusni opseg
    fa1 = (fc - 0.5)
    fa2 = (fc + 0.5)
    
    # predistorzija i prebacivanje f[Hz] -> w[rad/s]
    wp1 = 2 * fs * math.tan(fp1 * 2 * math.pi/fs/2)
    wp2 = 2 * fs * math.tan(fp2 * 2 * math.pi/fs/2)
    wa1 = 2 * fs * math.tan(fa1 * 2 * math.pi/fs/2)
    wa2 = 2 * fs * math.tan(fa2 * 2 * math.pi/fs/2)
    
    # propusni opseg
    B = wa2 - wa1
    # centralna ucestanost
    w0 = math.sqrt(wa1*wa2)
    
    if(wp2 > (w0**2)/wp1):
        wp2 = (w0**2) / wp1
    
    # normalizacija ucestanosti
    waN = 1
    wpN = (B * wp1) / (math.pow(w0, 2) - math.pow(wp1, 2))
    # faktor selektivnosti
    k = wpN/waN
    # faktor diskriminacije
    D = (math.pow(10, (0.1 * Aa)) - 1) / (math.pow(10, (0.1 * Ap)) - 1)
    # red funkcije prenosa
    N = math.ceil(math.acosh(math.sqrt(D)) / math.acosh(1/k))
    
    flag = 0
    while flag == 0:
        # odredjivanje nula i polova
        z, p, g = signal.cheb2ap(N, Aa)
        # odredjivanje koeficijenata polinoma na osnovu nula i polova
        bn = np.poly(z) * g
        an = np.poly(p)
        # transformacija u filtar nepropusnik opsega
        bh, ah = signal.lp2bs(bn, an, wo=w0, bw=B)
        # bilinearna transformacija
        # preslikavanje iz analognog u digitalni domen
        b, a = signal.bilinear(bh, ah, fs)
        # racunanje amplitudske karakteristike
        f, Hd = signal.freqz(b, a, worN=15000, fs=fs)
        hd = abs(Hd)
        
        # provera gabarita
        if (20*math.log10(max(hd[np.logical_and((f>=fa1), (f<=fa2))])) <= (-Aa + 0.001)) and (20*math.log10(min(hd[np.logical_or((f<fp1), (f>fp2))])) >= (-Ap - 0.001)):
            flag = 1
        else:
            N = N+1
            
    return b, a
```

```python
# projektovanje filtra pomocu funkcije power_line_noise_filter
fc = 60
Aap = 40
App = 0.5
bp, ap = power_line_noise_filter(fs, fc, Aap, App)
ecg_powerline_filtered = signal.lfilter(bp, ap, ecg_signal)
# vremenski oblik dobijenog signala
fig, ax = plt.subplots()
ax.plot(t, ecg_powerline_filtered)
ax.set(xlabel='t [s]', ylabel='EKG [mV]',
      title='Potpuno filtrirani EKG signal')
```

```python
# projektovanje filtra pomocu funkcije iirdesign
# PROVERI, NISAM SIGURAN KOJI PARAMETRI SE KORISTE
# baseline specifikacije
wab = 2 / Ts * math.tan(2*math.pi*fab/fs)
wpb = 2 / Ts * math.tan(2*math.pi*fpb/fs)

bbd, abd = signal.iirdesign(wab, wpb, gpass=0.5, gstop=40, ftype='cheby2', output='ba', fs=fs)
baseline_filtered = signal.lfilter(bbd, abd, ecg_signal)
fig, ax = plt.subplots()
ax.plot(t, baseline_filtered)
ax.set(xlabel='t [s]', ylabel='EKG [mV]',
      title='Filtrirani EKG signal')
```

# Deo 2: Implementacija FIR filtara za filtriranje nepoznatog zvučnog signala

```python
# ucitavanje nefiltriranog signala
fsound, sound = wav.read('sound_corrupted.wav')

# spektogram nefiltriranog signala
fig, ax = plt.subplots()
pxx,  freq, t, cax = ax.specgram(sound, NFFT=1500, noverlap=500, Fs=fsound)
ax.set(xlabel='Time [s]', ylabel='Frequency [Hz]',
       title='Spektogram nefiltriranog signala')

Audio(sound, rate=fsound)
```

```python
# metoda ogranicavanja impulsnog odziva
# Kajzerova prozorska funkcija

# NF filtar
def low_pass(fp, fa, Ap, Aa, fs):
    wp = 2 * math.pi * fp / fs
    wa = 2 * math.pi * fa / fs
    
    Bt = wa - wp
    wc = (wa + wp) / 2
    
    delta = min((math.pow(10, (0.05*Ap)) - 1)/(math.pow(10, (0.05*Ap)+1)), math.pow(10, (-0.05*Aa)))
    
    if delta != math.pow(10, -0.05*Aa):
        Aa = -20 * math.log10(delta)
        
    if Aa < 21:
        beta = 0
    elif Aa <= 50:
        beta = 0.5842 * (Aa - 21)**(0.4) + 0.07886 * (Aa - 21)
    else:
        beta = 0.1102 * (Aa - 8.7)
    
    if Aa <= 21:
        D = 0.9222
    else:
        D = (Aa - 7.95) / 14.36
        
    temp = (2 * math.pi * D) / Bt
    M = math.ceil(temp + 1)
    n1 = -(M - 1) / 2
    n2 = (M - 1) / 2
    n = np.arange(n1, n2) + math.pow(10, -9)
    
    hd_temp = np.sin(wc*n) / (n * math.pi)
    
    if M % 2 == 1:
        hd_temp[int((M+1)/2)] = wc / math.pi
    
    hd = hd_temp * np.transpose(signal.kaiser(M-1, beta))
    
    N = M - 1
    
    return hd, N
```

```python
# band stop filtar
def band_stop_filter(fp1, fp2, fa1, fa2, Ap, Aa, fs):
    Wp1 = 2 * math.pi * fp1 / fs
    Wp2 = 2 * math.pi * fp2 / fs
    Wa1 = 2 * math.pi * fa1 / fs
    Wa2 = 2 * math.pi * fa2 / fs
    
    Bt = min((Wa1 - Wp1), (Wp2 - Wa2))
    # opseg koji se filtrira
    Wc1 = Wp1 + Bt / 2
    Wc2 = Wp2 - Bt / 2
    
    delta = min(((math.pow(10, (0.05*Ap)-1))/math.pow(10, (0.05*Ap))+1), (math.pow(10, (-0.05*Aa))))
    
    if delta != math.pow(10, (-0.05*Aa)):
        Aa = - 20 * math.log10(delta)
    
    if Aa < 21:
        beta = 0
    elif Aa <= 50:
        beta = 0.5842 * (Aa - 21)**(0.4) + 0.07886 * (Aa - 21)
    else:
        beta = 0.1102 * (Aa - 8.7)
        
    if Aa <= 21:
        D = 0.9222
    else:
        D = (Aa - 7.95) / 14.36
        
    temp = (2 * math.pi * D) / Bt
    M = math.ceil(temp + 1)
    n1 = -(M - 1) / 2
    n2 = (M - 1) / 2
    n = np.arange(n1, n2) + math.pow(10, -9)
    
    # iz furijeovog reda idealne funkcije prenosa
    hd_temp = (-np.sin(n*Wc2) + np.sin(n*Wc1) + np.sin(n*math.pi)) / (n*math.pi)
    
    if M % 2 == 1:
        hd_temp[int((M+1)/2)] = (Wc1 - Wc2) / math.pi + 1
        
    hd =hd_temp * np.transpose(signal.kaiser(M-1, beta)) 
    
    N = M - 1
    
    return hd, N
```

```python
# low pass filter
fp = 6300
fa = 6500
Ap = 1

# prvi band stop filtar
fp11 = 1200
fp21 = 1700
fa11 = 1300
fa21 = 1600

# drugi band stop filtar
fp12 = 2400
fp22 = 3100
fa12 = 2500
fa22 = 3000

fig, axes = plt.subplots(ncols = 2, nrows = 7, figsize = (20,30), dpi = 100)
ax = axes.ravel()

i = 0
for Aa in np.arange(50,120, 10):
    fir_filter_low_pass, N = low_pass(fp, fa, Ap, Aa, fsound)
    sound_low_pass = signal.lfilter(fir_filter_low_pass, 1, sound)

    fir_filter_band_stop_1, N = band_stop_filter(fp11, fp21, fa11, fa21, Ap, Aa, fsound)
    sound_band_stop_1 = signal.lfilter(fir_filter_band_stop_1, 1, sound_low_pass)

    fir_filter_band_stop_2, N = band_stop_filter(fp12, fp22, fa12, fa22, Ap, Aa, fsound)
    sound_band_stop_2 = signal.lfilter(fir_filter_band_stop_2, 1, sound_band_stop_1)

    # spektogram signala nakon NF filtra
    ax[i].specgram(sound_low_pass, NFFT=4096, noverlap=500, Fs=fsound)
    ax[i].set_xlabel('Time [s]')
    ax[i].set_ylabel('Frequency [Hz]')
    ax[i].set_title('Spektogram signala nakon NF filtra - Aa = ' + str(Aa))

    # spektogram signala nakon band stop filtara
    ax[i+1].specgram(sound_band_stop_2, NFFT=4096, noverlap=500, Fs=fsound)
    ax[i+1].set_xlabel('Time [s]')
    ax[i+1].set_ylabel('Frequency [Hz]')
    ax[i+1].set_title('Spektogram filtriranog signala - Aa = ' + str(Aa))
    
    i += 2
    
plt.tight_layout()
plt.show()
```

# Deo 3: Implementacija direktne kanonicke realizacije IIR filtra

```python
# direknta II realizacija IIR filtra 

def iirDirectCan(b, a, x):
    lenb = len(b)
    lena = len(a)
    # dopuni nulama kraci niz
    if lena < lenb:
        a[end:end+lenb-lena] = 0
    else:
        b[end:end+lena-lenb] = 0
        
    # pomocni signali
    # matrice kolona
    x_temp = np.zeros(lenb, 1)
    y_temp = np.zeros(lena, 1)
    
    lenx = len(x)
    y = np.zeros(1, lenx)
    
    for i in range(len(x)):
        x_temp = [x[i], x_temp[0:end]]
        y_temp = [b*x_temp, y_temp[0:end]]
        
        y[i] = (-a) * y_temp
        y_temp[0] = y[i]
    
    return y
```

# Deo 4: Implementacija kaskadne realizacije IIR filtra

```python
def iirDirectCanCascade(b, a, x):
    #broj kaskada
    N = len(a)
    mr = np.zeros((3, N))
    izl = np.zeros((1, N-1))
    y = np.zeros((1, len(x)))

    for i in range(len(x)):
        izl[0] = x[i]
        mr[1:] = mr[0:-1]
        
        for kask in range(N):
            mr[0, kask] = izl[kask] - a[kask, 1:3] * mr[1: , kask]
            print('asd')
            izl[kask+1] = np.cross(b[kask, : ], mr[ : , kask])
        y[i] = izl[-1]
    return y
```

```python
fs = 360
t = np.arange(1, 360) / fs

f1 = 10
f2 = 60
f3 = 120

x = np.sin(2*math.pi*f1*t) + np.sin(2*math.pi*f2*t) + 2 * np.sin(2*math.pi*f3*t)

b, a = power_line_noise_filter(fs, f2, 30, 0.5)
sos = signal.tf2sos(b, a)
b = sos[:, 0:3]
a = sos[:, 3:6]
y = iirDirectCanCascade(b, a, x)
```
