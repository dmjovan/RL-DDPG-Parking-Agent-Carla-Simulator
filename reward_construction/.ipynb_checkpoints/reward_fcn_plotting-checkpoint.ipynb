## Importing necessary libriaries

```python
import numpy as np
import matplotlib.pyplot as plt
import os 
```

## Definitions of Reward function Rt 

```python
def reward_fcn(d, theta=None, Rmax=10, dmax=26, d_val_1=5, select_fcn='distance_exponential'):
    
    angle_penalty = abs(np.cos(np.deg2rad(theta))) if theta != None else 1.0
    
    if d >= dmax :
        Rt = -5*Rmax
        return Rt
    
    if select_fcn == 'distance_exponential':
        
        alpha = d_val_1/(-np.log(1.0/Rmax))
        
        Rt = Rmax*np.exp(-(d/alpha))*angle_penalty
        
    elif select_fcn == 'distance_linear':

        if d >= d_val_1 and d < dmax :
            Rt = (-1.0/(dmax-d_val_1))*d + dmax/(dmax-d_val_1)
            
        elif d >= 0 and d < d_val_1:
            Rt = (1-Rmax)*d/float(d_val_1) + Rmax
            
        Rt *= angle_penalty
            
    return Rt    
```

## Defining parameters for this analysis 

```python
num_points = 1000

d = np.linspace(0,25,num_points)

reward_dict = {'distance_linear': ['Part-by-part linear $R\mathrm{t}(d)$','red'],
               'distance_exponential': ['Exponential $R\mathrm{t}(d)$','green']}

all_rewards = list(reward_dict.keys())
```

## Distance influence on reward function Rt

```python
Rt = np.zeros((len(all_rewards), num_points))

fig = plt.figure(figsize=(10,10), dpi=100)

for i in range(len(all_rewards)):    
    for j in range(num_points):
        Rt[i,j] = reward_fcn(d[j], select_fcn=all_rewards[i])
        
    plt.plot(d, Rt[i], color=reward_dict[all_rewards[i]][1], linewidth=1.5, label=reward_dict[all_rewards[i]][0])

plt.title('Constructed Reward function $R\mathrm{t}(d)$')
plt.xlabel('d [m]')
plt.ylabel('$R\mathrm{t}(d)$')
plt.legend()
plt.grid()
fig.savefig(os.getcwd()+'/reward_plot_distance.png', facecolor = 'white', bbox_inches='tight')
```

## Distance an angle influence on reward function Rt

```python
theta = np.arange(0,180,15)
Rt_ = np.zeros((len(all_rewards), num_points, num_points))

for i in range(len(all_rewards)):    
    for j in range(num_points):
        for k in range(theta.shape[0]):
            Rt_[i,j,k] = reward_fcn(d[j], theta=theta[k], select_fcn=all_rewards[i])
```

```python
fig, axes = plt.subplots(ncols = 2, nrows = 1, figsize=(14,10), dpi=100)
ax = axes.ravel()

for i in range(len(all_rewards)):  
    for j in range(theta.shape[0]):
        ax[i].plot(d, Rt_[i,:,j], linewidth=1.5, label=str(theta[j])+'\N{DEGREE SIGN}')
        
    ax[i].set_title(reward_dict[all_rewards[i]][0])
    ax[i].set_xlabel('d [m]')
    ax[i].set_ylabel('$R\mathrm{t}(d, theta)$')
    ax[i].legend()
    ax[i].grid()
        
plt.tight_layout()
fig.savefig(os.getcwd()+'/reward_plot_distance_and_angle.png', facecolor = 'white', bbox_inches='tight')
```
